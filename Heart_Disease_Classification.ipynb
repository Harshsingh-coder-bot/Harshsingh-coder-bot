```python
#train_test-split
from sklearn.model_selection import train_test_split #sklearn is a library in pyhton we are useing train_test_split
np.random.seed(42) #split data in random seed
X_train,X_test, Y_train,Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
X_train
```

```python
Y_train
```

```python
from sklearn import linear_model as lim
from sklearn import neighbors as nei
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
```

```python
%%time
models={"LogisticRegression":lim.LogisticRegression(),
         "KNN":nei.KNeighborsClassifier(),
        "Random Forest":RandomForestClassifier(),
        "Ridge Classification":lim.RidgeClassifier(),
        "SVM":svm.SVC()
       }
def scoreModel(models,train_x,test_x,train_y,test_y): #scoremodel is a function created
    np.random.seed(42)
    model_scores={}
    for name, model in models.items():
        model.fit(train_x, train_y) #model is trained
        model_scores[name]=model.score(test_x,test_y)
    return model_scores

```

```python
model_scores=scoreModel(models,X_train,X_test,Y_train,Y_test)
model_scores
```

```python
model_compare=pd.DataFrame(model_scores, index=["accuracy"])
```

```python
model_compare.T.plot.bar()
```

### Hyperparameter Tuning

```python
#Tuning the Logistic regression
train_scores=[]
test_scores=[]
LR= lim.LogisticRegression()
qwe= range(50,500,5)
for i in qwe:
    LR.set_params( max_iter=i)
    LR.fit(X_train,Y_train)
    train_scores.append(LR.score(X_train,Y_train))
    test_scores.append(LR.score(X_test,Y_test))
train_scores
test_scores
```

```python
qwe= range(50,500,5)
fig, ax=plt.subplots(figsize=(18,12))
ax.plot(qwe, train_scores ,label="train scores")
ax.plot(qwe, test_scores ,label="test scores")
plt.xlabel('number of iterations')
plt.ylabel('Model Scores')
plt.legend()
```

### Using RandomizedSearchgrid

```python
reg_grid={'C': np.logspace(-4,4,20),
          'solver': ['lbfgs']}
```

```python
from sklearn.model_selection import RandomizedSearchCV
np.random.seed(42)
rsg=RandomizedSearchCV(lim.LogisticRegression(),
                      param_distributions=reg_grid,
                      cv=5,
                      n_iter=10,
                      verbose=True)
rsg.fit(X_train,Y_train) #training the model
```

```python
rsg.best_params_
```

```python
rsg.score(X_test,Y_test)
```

```python
y_preds=rsg.predict(X_test)
```
